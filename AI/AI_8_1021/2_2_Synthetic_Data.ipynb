# HTTPX

- Python ìš© HTTP í´ë¼ì´ì–¸íŠ¸
- ë¹„ë™ê¸° í†µì‹ ì„ ì§€ì›
  - ì—¬ëŸ¬ API ìš”ì²­ì„ ë³´ë‚¼ ë•Œ í•˜ë‚˜ì˜ ìš”ì²­ì´ ëë‚  ë•Œ ê¹Œì§€ ê¸°ë‹¤ë¦¬ì§€ ì•Šê³  ë™ì‹œì— ë³‘ë ¬ì ìœ¼ë¡œ ì²˜ë¦¬


```python
import httpx
import asyncio

# asyncê°€ ë¶™ìœ¼ë©´ ë¹„ë™ê¸° ì‘ì—…ì„ ìˆ˜í–‰í•˜ëŠ” í•¨ìˆ˜ì„ì„ ë‚˜íƒ€ëƒ„
async def call_chat_completion(url, headers, payload):
    print('   call_chat_completion ì‹œì‘')
    async with httpx.AsyncClient(timeout=30.0) as client:
        # await: ë’¤ì— ì‘ì„±ëœ ë¹„ë™ê¸° ì‘ì—…ì´ ì¢…ë£Œê°€ ë ë•Œê¹Œì§€ ëŒ€ê¸°
        # ë§Œì•½ ë‹¤ë¥¸ ë¹„ë™ê¸° ì‘ì—…ì´ ì˜ˆì •ë˜ì–´ ìˆë‹¤ë©´ í•´ë‹¹ ì‘ì—…ì„ ìˆ˜í–‰í•˜ê¸° ì‹œì‘
        response = await client.post(url, headers=headers, json=payload)
        # HTTP ì˜¤ë¥˜ê°€ ë°œìƒí•˜ë©´ ì˜ˆì™¸ë¥¼ ë°œìƒì‹œí‚¤ë„ë¡
        response.raise_for_status()
        data = response.json()

        return data['choices'][0]['message']['content']

```

```python
# ì—¬ëŸ¬ ë¹„ë™ê¸° ì‘ì—…ì„ ì‹¤í–‰í•˜ê³  ê²°ê³¼ë¥¼ ì²˜ë¦¬í•˜ëŠ” ë©”ì¸ í•¨ìˆ˜.
async def request(tasks):
    print('ì´ì œ ê° ìš”ì²­ ì‹¤í–‰ ì‹œì‘')
    # asyncio.gather(*tasks): ë¦¬ìŠ¤íŠ¸ì— ë‹´ê¸´ ëª¨ë“  ë¹„ë™ê¸° ì‘ì—…ì„ ë™ì‹œì— ì‹¤í–‰í•˜ê³ ,
    # ëª¨ë“  ì‘ì—…ì´ ì™„ë£Œë  ë•Œê¹Œì§€ ê¸°ë‹¤ë¦° í›„ ê²°ê³¼ë¥¼ ë¦¬ìŠ¤íŠ¸ë¡œ ëª¨ì•„ì„œ ë°˜í™˜.
    results = await asyncio.gather(*tasks)
    print('ëª¨ë“  ìš”ì²­ ì™„ë£Œ')
    print()

    # ì™„ë£Œëœ ê²°ê³¼ë¥¼ í•˜ë‚˜ì”© ì¶œë ¥.
    for i, res in enumerate(results, 1):
        print(f'Response {i}')
        print(res)
        print('=' * 20)


# API ì—”ë“œí¬ì¸íŠ¸ ë° ì¸ì¦ í—¤ë” ì„¤ì •.
url = "https://api.upstage.ai/v1/chat/completions"
headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {UPSTAGE_API_KEY}",
}

# ë™ì‹œì— ë³´ë‚¼ ì—¬ëŸ¬ ê°œì˜ í”„ë¡¬í”„íŠ¸ë¥¼ ë¦¬ìŠ¤íŠ¸ë¡œ ì¤€ë¹„.
prompts = [
    "ë†ë‹´ í•˜ë‚˜ë§Œ í•´ ì¤˜",
    "í”„ë‘ìŠ¤ì˜ ìˆ˜ë„ëŠ” ì–´ë””ì•¼?",
]

# ì‹¤í–‰í•  ë¹„ë™ê¸° ì‘ì—…ë“¤ì„ ë‹´ì„ ë¦¬ìŠ¤íŠ¸.
tasks = []
# ê° í”„ë¡¬í”„íŠ¸ì— ëŒ€í•´ API ìš”ì²­ í˜ì´ë¡œë“œë¥¼ ìƒì„±.
for prompt in prompts:
    payload = {
        "model": "solar-pro2",
        "messages": [
            {
                "role": "user",
                "content": prompt,
            }
        ],
        # stream=False: ì „ì²´ ì‘ë‹µì„ í•œ ë²ˆì— ë°›ë„ë¡ ì„¤ì •.
        "stream": False,
    }
    # call_chat_completion í•¨ìˆ˜ë¥¼ í˜¸ì¶œí•˜ì—¬ ë¹„ë™ê¸° ì‘ì—…(Task) ê°ì²´ë¥¼ ìƒì„±í•˜ê³  ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€.
    # 'await'ê°€ ì—†ìœ¼ë¯€ë¡œ í•¨ìˆ˜ê°€ ë°”ë¡œ ì‹¤í–‰ë˜ì§€ ì•Šê³ , ì‹¤í–‰ ëŒ€ê¸° ìƒíƒœì˜ ê°ì²´ë§Œ ë§Œë“¤ì–´ì§.
    print('íƒœìŠ¤í¬ ìƒì„±')
    tasks.append(call_chat_completion(url, headers, payload))
    print('ì•„ì§ upstage API í˜¸ì¶œ ì „')

# ì¤€ë¹„ëœ ëª¨ë“  íƒœìŠ¤í¬ë¥¼ ë™ì‹œì— ì‹¤í–‰.
await request(tasks)
```

## ì‘ë‹µì„ JSONìœ¼ë¡œ ê³ ì •

```python
import json

# LLMì˜ ì‘ë‹µì„ êµ¬ì¡°í™”ëœ JSON í˜•ì‹ìœ¼ë¡œ ê°•ì œí•˜ê¸° ìœ„í•œ ì„¤ì •.
response_format = {
    "type": "json_schema",
    "json_schema": {
        "name": "ìˆ˜ë„ ì •ë³´",
        "strict": True,
        "schema": {
            "type": "object",
            "properties": {
                "capital": {"type": "string"},
                "translation": {"type": "string", "description": "ìˆ˜ë„ì˜ ì˜ì–´ ë²ˆì—­"},
            },
            "required": ["capital", "translation"],
        },
    },
}

# client.chat.completions.create ë©”ì„œë“œë¥¼ í˜¸ì¶œí•˜ì—¬ LLMì— ìš”ì²­.
response = client.chat.completions.create(
    model="solar-pro2",
    messages=[
        {
            "role": "user",
            "content": "í•œêµ­ì˜ ìˆ˜ë„ëŠ” ì–´ë””ì•¼?",
        }
    ],
    # ìœ„ì—ì„œ ì •ì˜í•œ JSON ìŠ¤í‚¤ë§ˆë¥¼ ì ìš©í•˜ì—¬ ì‘ë‹µ í˜•ì‹ì„ ê°•ì œ.
    response_format=response_format,
)

# LLMì˜ ì‘ë‹µì€ JSON í˜•ì‹ì˜ 'ë¬¸ìì—´'ì´ë¯€ë¡œ,
# json.loadsë¥¼ ì‚¬ìš©í•˜ì—¬ íŒŒì´ì¬ ë”•ì…”ë„ˆë¦¬ ê°ì²´ë¡œ ë³€í™˜.
print(response.choices[0].message.content)
structured_dictionary = json.loads(response.choices[0].message.content)

# ë”•ì…”ë„ˆë¦¬ë¡œ ë³€í™˜ëœ êµ¬ì¡°í™”ëœ ì‘ë‹µì„ ì¶œë ¥.
print("Structured Response:")
for key, value in structured_dictionary.items():
    print(f"{key}: {value}")
```

```python
# Upstage APIì™€ í†µì‹ í•˜ê¸° ìœ„í•´ openai ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ì„í¬íŠ¸í•©ë‹ˆë‹¤.
from openai import OpenAI
# LLMì˜ ì‘ë‹µ(JSON í˜•ì‹ì˜ ë¬¸ìì—´)ì„ íŒŒì´ì¬ ë”•ì…”ë„ˆë¦¬ë¡œ ë³€í™˜í•˜ê¸° ìœ„í•´ json ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ì„í¬íŠ¸í•©ë‹ˆë‹¤.
import json

# ì‹œìŠ¤í…œ í”„ë¡¬í”„íŠ¸: AI ëª¨ë¸ì˜ ì—­í• ê³¼ ê¸°ë³¸ í–‰ë™ ì§€ì¹¨ì„ ì •ì˜í•©ë‹ˆë‹¤.
SYSTEM_PROMPT = """
ë‹¹ì‹ ì€ ì„¸ìƒì˜ ëª¨ë“  ì˜í™”ë¥¼ ê¿°ëš«ê³  ìˆëŠ” ì˜í™” ì „ë¬¸ê°€ 'ì‹œë„¤ë§ˆìŠ¤í„°'ì…ë‹ˆë‹¤.
ì‚¬ìš©ìì˜ ìš”ì²­ì— ë§ì¶° ì˜í™”ë¥¼ ì¶”ì²œí•˜ëŠ” ì—­í• ì„ ë§¡ê³  ìˆìŠµë‹ˆë‹¤. ì˜í™”ëŠ” ë°˜ë“œì‹œ í•˜ë‚˜ë§Œ ì¶”ì²œí•©ë‹ˆë‹¤.
"""

# ì¶”ê°€ ê·œì¹™ í”„ë¡¬í”„íŠ¸: AI ëª¨ë¸ì˜ ë§íˆ¬ë‚˜ ë‹µë³€ ìŠ¤íƒ€ì¼ ë“± ì„¸ë¶€ ê·œì¹™ì„ ì •ì˜í•©ë‹ˆë‹¤.
RULE = """
ì¹œêµ¬ê°€ ì†Œê°œ í•´ì£¼ëŠ” ë“¯ ë¶€ë“œëŸ½ê³  ì¹œê·¼í•œ ë§íˆ¬ë¡œ ë‹µë³€í•©ë‹ˆë‹¤.
íŠ¹íˆ, recommended_reason í•­ëª©ì—ì„œëŠ” ì¹œêµ¬ê°€ ì—„ì²­ í˜¸ë“¤ê°‘ ë–¨ë“¯ì´ ì„¤ëª…í•´ ì£¼ì„¸ìš”.
"""

# ì‘ë‹µ í˜•ì‹ í”„ë¡¬í”„íŠ¸: LLMì˜ ë‹µë³€ì„ êµ¬ì¡°í™”ëœ JSON í˜•ì‹ìœ¼ë¡œ ê°•ì œí•˜ê¸° ìœ„í•œ ì„¤ì •ì…ë‹ˆë‹¤.
response_format = {
    "type": "json_schema",      # ì‘ë‹µ í˜•ì‹ì„ JSON ìŠ¤í‚¤ë§ˆë¡œ ì§€ì •í•©ë‹ˆë‹¤.
    "json_schema": {
        "name": "ì˜í™” ì¶”ì²œ",      # ìŠ¤í‚¤ë§ˆì˜ ì´ë¦„ì„ ì§€ì •í•©ë‹ˆë‹¤.
        "strict": True,          # ì—„ê²© ëª¨ë“œ: ìŠ¤í‚¤ë§ˆì— ë§ì§€ ì•ŠëŠ” ì‘ë‹µì´ ì˜¤ë©´ ì˜¤ë¥˜ë¥¼ ë°œìƒì‹œí‚µë‹ˆë‹¤.
        "schema": {
            "type": "object",    # ì‘ë‹µì˜ ìµœìƒìœ„ íƒ€ì…ì´ ê°ì²´(ë”•ì…”ë„ˆë¦¬)ì„ì„ ì§€ì •í•©ë‹ˆë‹¤.
            "properties": {      # ê°ì²´ì— í¬í•¨ë  ì†ì„±ë“¤ì„ ì •ì˜í•©ë‹ˆë‹¤.
                "movie_name": {"type": "string"},
                "year": {"type": "integer"},
                "reason": {"type": "string"},
                "description": {"type": "string", "description": "ì˜í™”ì— ëŒ€í•œ ì„¤ëª…"},
                "recommended_reason": {"type": "string", "description": "ì´ ì˜í™”ë¥¼ ì¶”ì²œí•˜ëŠ” ì¶”ê°€ ì´ìœ "}
            },
            # í•„ìˆ˜ì ìœ¼ë¡œ í¬í•¨ë˜ì–´ì•¼ í•  ì†ì„±ë“¤ì„ ì§€ì •í•©ë‹ˆë‹¤.
            "required": ["movie_name", "year", "reason", "description", "recommended_reason"]
        }
    }
}

# Upstage API í´ë¼ì´ì–¸íŠ¸ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.
client = OpenAI(
    api_key=UPSTAGE_API_KEY,
    base_url="https://api.upstage.ai/v1"
)

# client.chat.completions.create ë©”ì„œë“œë¥¼ í˜¸ì¶œí•˜ì—¬ LLMì— ìš”ì²­ì„ ë³´ëƒ…ë‹ˆë‹¤.
response = client.chat.completions.create(
    model="solar-pro2",
    messages=[
        {
            "role": "system",
            "content": SYSTEM_PROMPT
        },
        {
            "role": "system",
            "content": RULE
        },
        {
            "role": "user",
            "content": "ê³µí¬ ì˜í™”ë¥¼ ì¶”ì²œí•´ì¤˜"
        }
    ],
    response_format=response_format,  # ìœ„ì—ì„œ ì •ì˜í•œ JSON ì‘ë‹µ í˜•ì‹ì„ ì ìš©í•©ë‹ˆë‹¤.
    temperature=0.5,                  # ë‹¤ì–‘ì„±ì„ ìœ„í•´ temperatureë¥¼ 0.5ë¡œ ì„¤ì •í•©ë‹ˆë‹¤.
    max_tokens=1000,                  # ì‘ë‹µì˜ ìµœëŒ€ ê¸¸ì´ë¥¼ 1000 í† í°ìœ¼ë¡œ ì œí•œí•©ë‹ˆë‹¤.
    top_p=1.0,                        # top_pë¥¼ 1.0ìœ¼ë¡œ ì„¤ì •í•˜ì—¬ ëª¨ë“  ë‹¨ì–´ë¥¼ í›„ë³´ë¡œ ê³ ë ¤í•©ë‹ˆë‹¤.
    n=1,                              # 1ê°œì˜ ì‘ë‹µë§Œ ìƒì„±í•©ë‹ˆë‹¤.
    frequency_penalty=0.0,            # íŠ¹ì • ë‹¨ì–´ì˜ ë°˜ë³µì„ ì–µì œí•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.
    presence_penalty=0.0              # ìƒˆë¡œìš´ ì£¼ì œì˜ ë“±ì¥ì„ ì¥ë ¤í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.
)

# ì‘ë‹µ ê²°ê³¼ì—ì„œ ì‹¤ì œ í…ìŠ¤íŠ¸ ë‚´ìš©ë§Œ ì¶”ì¶œí•©ë‹ˆë‹¤.
output = response.choices[0].message.content
print(output)
```
