# 청킹 & 파싱

-  파싱(Parsing) :
    - LLM은 기본적으로 문자로 입력받음
    - PDF, 웹페이지, 이미지 같은 형식의 문서는 다시 순수한 텍스트와 메타데이터로 구성된 형식이 필요

- 청킹(Chunking)
    - Context Window: LLM이 한번에 처리할 수 있는 텍스트의 양
    - Context Window의 크기에 맞는 적은 양의 데이터로 나누는 작업

- 토크나이징
    - 토큰화 하는것
    - 긴 문장을 의미를 가진 단위로 나누는 과정

```python
import tiktoken

# 예시 텍스트
text = """
AI 온라인 서점입니다. 고객님의 소중한 문의에 감사드립니다.
배송 정책에 대해 안내해 드리겠습니다.
일반 도서의 경우 오후 3시 이전 주문 시 당일 발송됩니다.
주말 및 공휴일은 배송이 어렵습니다.
제주 및 도서 산간 지역은 추가 배송비가 발생할 수 있습니다.
주문 번호 order-123의 배송 상태를 조회하시려면 마이페이지에서 확인 부탁드립니다.
"""

# 모델에 따라 인코딩 방식이 다를 수 있음. 여기서는 'cl100k_base'를 사용.
encoding = tiktoken.get_encoding("cl100k_base")
tokens = encoding.encode(text)

print("--- 토크나이징 결과 ---")
# 원문 텍스트의 길이와 토큰의 갯수는 1:1로 대응되지 않음을 확인.
print(f"원문 텍스트 길이: {len(text)} 문자")
print(f"토큰 수: {len(tokens)} 토큰")
# 첫 10개 토큰 예시.
print(f"토큰 예시: {tokens[:10]}...")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# 여러가지 설정으로 진행
chunking_configs = [
    {"chunk_size": 100, "chunk_overlap": 0},
    {"chunk_size": 100, "chunk_overlap": 20},
    {"chunk_size": 50, "chunk_overlap": 0},
    {"chunk_size": 50, "chunk_overlap": 10},
]

print("\n--- 청킹 결과 비교 ---")

for config in chunking_configs:
    chunk_size = config["chunk_size"]
    chunk_overlap = config["chunk_overlap"]

    # 새로운 `RecursiveCharacterTextSplitter` 선언
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap,
        # 청크의 길이 기준은 `len` 함수로 설정.
        # 즉, 문자의 길이를 기준으로 판단.
        length_function=len,
        is_separator_regex=False,
    )

    # `RecursiveCharacterTextSplitter`로 청킹 진행
    chunks = text_splitter.create_documents([text])

    # 결과 출력
    print(f"\n[Chunk Size: {chunk_size}, Chunk Overlap: {chunk_overlap}]")
    print(f"생성된 청크 개수: {len(chunks)}")
    for i, chunk in enumerate(chunks):
        print(f"  청크 {i+1} (길이: {len(chunk.page_content)}):")
        print(f"  '{chunk.page_content[:50]}...'") # 각 청크의 앞부분만 출력
```

# LangChain

LLM Chain을 제공하는 라이브러리

## LLM Chain

1. 입력을 프롬프트에 전달
2. 모델의 출력을 받아 원하는 형식으로 처리

```python
# LLM Chain 구성하는 법.
# 1. llm 정의, 2. prompt 정의, 3. chain 정의, 4. chain 호출

# 1. define your favorate llm, solar
from langchain_upstage import ChatUpstage
from os import getenv

llm = ChatUpstage(upstage_api_key=getenv("UPSTAGE_API_KEY"))

# 2. define chat prompt
from langchain_core.prompts import ChatPromptTemplate # '대화' 형태로 prompt template 생성

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "모든 답변은 존댓말로 답변해줘."),

        # few-shot prompting
        ("human", "프랑스의 수도는 어딘지 알아?"),  # human request
        ("ai", "알고 있습니다. 파리입니다."),      # LLM response
        ("human", "일본의 수도는 어딘지 알아?"),
        ("ai",  "알고 있습니다. 도쿄입니다."),

        # User Query
        ("human", "그렇다면, 한국은?"),
    ]
)

# 3. define chain
from langchain_core.output_parsers import StrOutputParser #문자열(text, string)만 나오게 하는 출력 파서

# chain = prompt | llm  # without output parser

chain = prompt | llm | StrOutputParser() # with output parser

# 4. invoke the chain
c_result = chain.invoke({})
print(c_result)
```

# RAG

Retrieval-Augmented Generation

- 외부의 신뢰할 수 있는 정보원에서 관련 정보를 **검색**하여 그 정보를 **참고**해서 답변을 **생성**하도록 돕는 기술
- LLM 환각현상 해소 : RAG를 이용하면 답변의 근거가 되는 문서를 제공함으로서 신뢰성을 확보 가능
- 도메인 지식, 최신 지식 활용 : LLM이 학습되지 않은 전문 분야의 지식 또는 학습된 후 시점의 지식을 활용할 수 있음

## Vector Store

데이터를 벡터로 저장

- 주어진 자료를 청크 단위로 분해하여 Vector Store에 저장
- 주어진 질문과 가장 연관이 높은 데이터를 Vector Store에서 회수



```python
# 필요한 모듈 추가
from langchain_community.document_loaders import TextLoader
from langchain_community.document_loaders import PyMuPDFLoader


# 사용하는 모든 문서를 저장할 리스트
documents = []

# 찾은 각 PDF 파일에 대해 파싱을 수행.
for pdf_filepath in pdf_files:
    try:
        loader = PyMuPDFLoader(pdf_filepath)
        pages = loader.load()
        # 로드된 페이지를 documents에 추가
        documents.extend(pages)
        print(f"'{pdf_filepath}' 파일 로드 완료. {len(pages)} 페이지 로드됨.")
    except Exception as e:
        print(f"'{pdf_filepath}' 파일 로드 중 오류 발생: {e}")

# shipping_policy.txt 파일 위치 지정(현재 경로의 텍스트 파일).
shipping_policy_file_path = './shipping_policy.txt'

try:
    # 텍스트 파일의 위치를 바탕으로 `TextLoader` 생성
    loader = TextLoader(shipping_policy_file_path)

    # 문서 불러오기
    text_documents = loader.load()

    # 불러온 문서를 documents에 추가
    documents.extend(text_documents)
    print(f"'{shipping_policy_file_path}' 파일 로드 완료. {len(text_documents)} 문서 로드됨.")


except FileNotFoundError:
    print(f"Error: File not found at {shipping_policy_file_path}. Please ensure the file exists at this location.")
except Exception as e:
    print(f"An error occurred while loading {shipping_policy_file_path}: {e}")


# 모든 로드된 총 문서(PDF 페이지 + Text 문서) 정보 출력
print("\n--- 모든 파일에서 로드된 총 문서 정보 ---")
print(f"총 문서 개수: {len(documents)}")
if documents:
    print(f"첫 번째 문서 내용 (일부): {documents[0].page_content[:200]}...")
    print(f"첫 번째 문서 메타데이터: {documents[0].metadata}")
else:
    print("로드된 문서가 없습니다.")
print("-" * 20)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter


# RecursiveCharacterTextSplitter를 사용하여 문서 분할.
# 여기서는 예시로 설정하며, 실제 적용 시에는 데이터 특성에 맞게 조정 필요.
text_splitter = RecursiveCharacterTextSplitter(
    # 청크 최대 크기 (문자 단위)
    chunk_size=200,
    # 청크 간 겹치는 부분 (문자 단위)
    chunk_overlap=20,
    # 길이를 문자로 계산
    length_function=len,
    is_separator_regex=False,
)


chunks = text_splitter.split_documents(documents)

print("\n--- 분할된 청크 정보 ---")
print(f"생성된 청크 개수: {len(chunks)}")
for i, chunk in enumerate(chunks):
    print(f"  청크 {i+1} (길이: {len(chunk.page_content)}):")
    print(f"  '{chunk.page_content[:100]}...'") # 각 청크의 앞부분만 출력
    print(f"  메타데이터: {chunk.metadata}")
print("-" * 20)
```

# 청크를 Vector Store에 저장

```python
from langchain_community.vectorstores import Chroma
from langchain_upstage import UpstageEmbeddings

# UpstageEmbeddings 인스턴스 생성
# 텍스트, 이미지 등의 데이터를 벡터로 변환해주는 Upstage의 임베딩 모델을 LangChain과 연결해주는 객체
embeddings = UpstageEmbeddings(model="embedding-query")

# Chroma Vector Store 생성
# 이전 단계에서 만든 `chunks` 활용
# chunks = text_splitter.split_documents(documents)
vectorstore = Chroma.from_documents(documents=chunks, embedding=embeddings)

print("\n--- Vector Store 생성 완료 ---")
# 확인을 위해 Vector Store에 있는 문서 수 출력
try:
    print(f"Vector Store에 저장된 문서 개수: {vectorstore._collection.count()}")
except Exception as e:
    print(f"Vector Store 문서 개수 확인 중 오류 발생: {e}")


# Vector Store를 검색기로 변환
retriever = vectorstore.as_retriever()

print("--- Retriever 생성 완료 ---")
```

### Naive RAG 수행

```python
from langchain_core.runnables import RunnablePassthrough
from langchain_core.prompts import ChatPromptTemplate
from langchain_upstage import ChatUpstage
from operator import itemgetter

# 이전 단계에서 만든 llm과 retriever를 활용
# llm = ChatUpstage()
# retriever = vectorstore.as_retriever


# RAG 체인을 위한 프롬프트 템플릿 정의
# 검색된 컨텍스트와 사용자 질문을 위한 placeholder 포함.
rag_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "너는 AI 온라인 서점의 고객 서비스 에이전트야.\n모든 답변은 존댓말로 답변해줘.\n다음 컨텍스트를 참고해서 질문에 답변해:\n\n{context}"),
        ("human", "{question}"),
    ]
)

# 검색된 문서 목록을 하나의 문자열로 변환하는 함수
def format_docs(docs):
    """문서 목록을 하나의 문자열로 포맷합니다."""
    return "\n\n".join(doc.page_content for doc in docs)


# LCEL(LangChain Expression Language)을 사용하여 Naive RAG 체인 구성
# 1. 사용자 질문을 입력받음 ('question' 키를 가진 딕셔너리 형태).
# 2. itemgetter('question')을 사용하여 `질문 문자열` 추출
# 3. `질문 문자열`을 retriever에게 전달하여 관련 문서 조회
# 4. format_docs를 사용하여 검색된 문서의 형식 변환
# 5. 포맷된 컨텍스트와 원본 질문 (RunnablePassthrough 사용)을 rag_prompt에게 전달
# 6. 만들어진 프롬프트를 llm에 보내 최종 답변을 생성
rag_chain = (
    RunnablePassthrough.assign(context=itemgetter("question") | retriever | format_docs)
    | rag_prompt
    | llm
)

print("--- Naive RAG 체인 구성 완료 ---")
# 확인을 위해 체인 구조 출력
print(rag_chain)


# 테스트 질문으로 RAG 체인 실행
test_question = "주말에도 배송되나요?"

print(f"\n--- 질문 '{test_question}'으로 RAG 체인 테스트 중 ---")

# 테스트 질문으로 체인 호출(ChatMessage 객체)
try:
    response = rag_chain.invoke({"question": test_question})
    print("\n--- RAG 체인 응답 ---")
    print(response.content)
    print("-" * 20)
except Exception as e:
    print(f"\n--- RAG 체인 실행 오류 ---")
    print(f"오류: {e}")
    print("'retriever'와 'llm'이 이전 셀에 정의되어 있는지 확인하십시오.")
    print("-" * 20)

# 다른 질문
# test_question_2 = "제주도 추가 배송비가 있나요?"
# print(f"\n--- 질문 '{test_question_2}'으로 RAG 체인 테스트 중 ---")
# try:
#     response_2 = rag_chain.invoke({"question": test_question_2})
#     print("\n--- RAG 체인 응답 ---")
#     print(response_2.content)
#     print("-" * 20)
# except Exception as e:
#     print(f"\n--- RAG 체인 실행 오류 ---")
#     print(f"오류: {e}")
#     print("'retriever'와 'llm'이 이전 셀에 정의되어 있는지 확인하십시오.")
#     print("-" * 20)
```

1. 최신 정보에 접근 및 활용'
2. 정확하고 신뢰성 있는 정보를 제공

### Reitriever를 Tool로 변환

AI Agent가 특정 작업을 수행하기 위해,

```python
# langchain.tools 모듈에서 retriever 도구를 생성하는 함수 추가
from langchain.tools.retriever import create_retriever_tool


# Tool을 보관할 리스트 생성
agent_tools = []

# retriever tool 생성
if retriever:
    shipping_policy_retriever_tool = create_retriever_tool(
        retriever,
        # Agent Tool 이름
        "shipping_policy_search_tool",
        # Agent Tool 설명 (English is recommended)
        "Searches and returns information about the AI online bookstore's shipping policy."
    )

    agent_tools.append(shipping_policy_retriever_tool)

    print("\n--- 배송 정책 검색 도구 생성 완료 ---")
    print(f"생성된 도구: {agent_tools}")
    print("-" * 20)
else:
    print("오류: 'retriever' 변수가 생성되지 않았습니다. Vector Store 및 Retriever 생성 단계를 먼저 실행해주세요.")
```

### 배송 상태 조회 Tool 생성

```python
# 함수를 Tool로 변환하는 tool 데코레이터 추가
from langchain_core.tools import tool

# @tool이 추가되면 LLM이 사용할 수 있는 Tool이 된다.
@tool
def get_order_status(order_id: str) -> str:
    # docstring의 형태로 Tool의 Description 정의
    """주문 번호를 입력받아 해당 주문의 배송 상태를 반환합니다.
    알 수 없는 주문 번호의 경우, 해당 주문을 찾을 수 없음을 알립니다.
    """
    # 실제 시스템에서는 데이터베이스 조회 등을 실행하여 조회.
    # 여기서는 예시를 위해 더미 데이터를 사용.
    order_statuses = {
        "order-123": "상품 준비 중",
        "order-456": "배송 중",
        "order-789": "배송 완료",
    }

    status = order_statuses.get(order_id, f"주문 번호 {order_id}를 찾을 수 없습니다.")
    return status

print("--- 주문 조회 도구 생성 완료 ---")
print(f"생성된 도구 이름: {get_order_status.name}")
print(f"생성된 도구 설명: {get_order_status.description}")
print("-" * 20)

# 생성된 도구를 리스트에 추가 (이전 단계에서 만들었던 리스트에 추가)
# agent_tools = []
agent_tools.append(get_order_status)

# 도구 사용 예시 (실제 에이전트가 호출하는 방식)
# print(get_order_status.invoke("order-123"))
# print(get_order_status.invoke("order-999"))
```

```python
print("--- 현재 에이전트 도구 목록 ---")
for tool in agent_tools:
    print(tool.name)
```

### 고객 인사 응대 Tool 생성

```python
from langchain.tools import tool

@tool
def greet_customer(query: str) -> str:
    """고객의 인삿말(예: 안녕, 안녕하세요)이 포함되어 있으면, 정해진 환영 메시지를 반환합니다."""
    greeting_keywords = ["안녕", "안녕하세요", "하이", "반가워요"]
    if any(keyword in query for keyword in greeting_keywords):
        return "안녕하세요! AI 온라인 서점입니다. 무엇을 도와드릴까요?"
    else:
        # 인삿말이 아닌 경우 이 Tool을 사용하지 않도록 빈 문자열 또는 None을 반환
        return ""

print("--- 인사 도구 생성 완료 ---")
print(f"생성된 도구 이름: {greet_customer.name}")
print(f"생성된 도구 설명: {greet_customer.description}")
print("-" * 20)

# 생성된 도구를 기존 에이전트 도구 리스트에 추가
agent_tools.append(greet_customer)
```

```python
print("--- 현재 에이전트 도구 목록 ---")
for tool in agent_tools:
    print(tool.name)
```
